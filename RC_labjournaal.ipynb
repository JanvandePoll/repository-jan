````python
#ANTWOORD
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# Gegevens
f = 5
T = 1/f
A = 2.5
Vd = 1.7
R = 220

t = np.linspace(0, T, 2000)
vs = A * np.sin(2*np.pi*f*t)

# spanning over R
vR = np.maximum(vs - Vd, 0)

s = 0
for i in range(len(t)):
    if vs [i]>1.7:
        s += 1
percentage = s /2000 *100
print("gevonden percentage is %.2f procent " % (percentage) )

# Plotten
plt.plot(t, vs, label="Bronspanning Vs(t)")
plt.plot(t, vR, label="Spanning over R (vR)", linewidth=2)
plt.axhline(1.7, color='gray', linestyle='--', label="LED drempel")
plt.legend()
````

## RC-circuit vervolg: RC-tijd en het low-pass filter. (∼ 45 min.)

Een condensator bestaat dus uit twee platen waar je lading op kunt slaan. Als er een gelijkspanning over de condensator gezet wordt, zal deze langzaam ‘vollopen’. Bij de start van het opladen zal de spanning met name over de weerstand staan, maar deze spanning verschuift langzaam naar de condensator.

Een karakteristieke tijd voor het vollopen is de RC-tijd $τ = RC$. De RC-tijd is gelijk aan de tijd die nodig is om de spanning over de condensator gelijk aan $63.2\%$ ($1 − e^{−1}$) van de bronspanning te laten zijn en volgt op logischer wijze uit de formule voor de spanning over de condensator: 

$$V(t) = V_0 \cdot e^{-\frac{t}{RC}}$$

Bij een wisselspanning zal de condensator opladen en vervolgens weer ontladen. Als de frequentie van de wisselspanning heel hoog is, zal de condensator zich gedragen als een draad zonder weerstand. De stroomsterkte wordt dan met name bepaald door de weerstandswaarde en de spanning over de condensator zal klein zijn. Dat de spanning over de condensator sterk afneemt is goed te zien vanaf een bepaalde frequentie, de cutoff frequency: 

$$f_{cutoff} = \frac{1}{2πRC}$$

Beide eigenschappen van een RC-circuit onderzoeken we in dit deel van het practicum.

### Low-pass filter toegepast
Een Low-Pass Filter kan heel handig zijn om ruis uit een signaal te filteren. We gaan een signaal ontwerpen dat bestaat uit de som van een laagfrequente sinus (het signaal waar we in geintresseerd zijn) en een hoogfrequente sinus (de ruis). De ruis kan uit het signaal gefiltered worden met het low-pass filter uit de vorige opdracht. In deze opdracht gaan we onderzoeker hoe goed dit filter werkt.
  
Om dit signaal te genereren, moeten we gebruik maken van de geadvanceerde opties van de functiegenerator. We hebben in dit geval de SDG1032X arbitrary waveform genetor ter beschikking die de mogelijkheid beidt om zelf een signaal te definiëren. Deze _aritrary wavefrom_ maken we in python en sturen we vervolgens ook naar de functiegenerator met python. Gebruik hiervoor de computer in de studio classroom die het dichtst bij de caddy met de apparatuur staat. Deze computer is namelijk via USB aangesloten op je functiegenerator.  
  
De benodigde code is gegeven:

````python
import pyvisa   # pyvisa regelt de communicatie met de functiegenerator
import struct   # struct gebruiken we om het juiste databestand the genereren

# Constants:
SIZE_INT16 = 2**16
````

De onderstaande code maakt verbinding met de functiegenerator. Zorg dat de functiegenerator aan staat. 

````python
rm = pyvisa.ResourceManager()
resources = rm.list_resources()
my_instrument = None

for device in resources:
    if device.find('SDG')>-1: # -1 = apparaat niet gevonden
        my_instrument = rm.open_resource(device)
        print('Signal generator found!', my_instrument.query('*IDN?'), 'conected at', device)

if my_instrument is None:
    print('Signal generator not found. Make sure it is turned on and connected, or select manually.')
````

We kunnen de functiegenerator nu aanroepen via _my\_instrument_.  
  
We definiëren een aantal hulp functies. De functie sin($t,f$) evalueert een sinus met frequentie $f$. De functie make_sine_sum($f_1, f_2$) telt twee sinussen van frequentie $f_1$ en $f_2$ bij elkaar op. Om een continue signaal te genereren moeten $f_1$ en $f_2$ een gemeenschappelijke deler hebben. De functie create_waveform() zet het signaal om naar een bestand dat de functiegerenator uit kan lezen.

````python
def sin(t, f):
    """ Evalueert een sinus van frequentie f op tijdstip t. """
    return np.sin(2*np.pi * t*f)

def make_sine_sum(f1, f2, ampl1=1, ampl2=1, N=1000):
    fsum = np.gcd(f1, f2) # important, f1 and f2 should have a common factor.
    t = np.linspace(0, 1/fsum, N) # 
    wave = ampl1*sin(t,f1) + ampl2*sin(t,f2)
    return t, wave

def create_waveform(wave, filename, amplitude):
    wave_data = [int(amplitude*value) % SIZE_INT16 for value in wave]
    with open(filename, 'wb') as f:
        for value in wave_data:
            packed_data = struct.pack('<H', value)  # '<H' for 2-byte unsigned short, little-endian
            f.write(packed_data)
````

Maak nu het signaal en plot het. Herken je de sinus met ruis? Hoeveel datapunten $N$ heeft het signaal nodig?

````python
t, wave = make_sine_sum(f1=500, f2=15000, ampl1=1, ampl2=0.1, N=1000) # original values: 500, 15000, 1, 0.1, 1000
plt.figure()
plt.plot(t, wave)
plt.xlabel("$t$ [s]")
plt.ylabel("signal [-]")
plt.xticks()
plt.show()
````

We gaan het signaal nu verzenden naar de functiegenerator. Om te testen of dit goed werkt kun je de fucntiegenerator het best direct verbinden met een BNC kabel naar de oscilloscoop.  
  
We moeten naast de vorm van het signaal ook de gewenste amplitude en frequentie meegeven. Komt de gemeten frequentie overeen met de eerder gespecificeerde frequentie van 500Hz?

````python
create_waveform(wave, "wave1.bin", amplitude=2**14) # original: wave, "wave1.bin", 2**14

with open('wave1.bin', 'rb') as f:
    data = f.read().decode('latin1')

my_instrument.write('C1:OUTP OFF, LOAD,HZ') # turn off output while making adjustments
amplitude_sg = "2.0" # Maximum amplitude, original: "2.0"
frequency_sg = "800.0" # frequency of the signal, original: "800.0"
my_instrument.write("C1:WVDT WVNM,wave1,FREQ,%s,AMPL,%s,OFST,0.0,PHASE,0.0,WAVEDATA,%s" % (frequency_sg, amplitude_sg, data), encoding='latin1') # set waveform data of wave1
my_instrument.write('C1:ArbWaVe NAME, wave1') # set mode to output arbitrary waveform wave1
my_instrument.write('C1:OUTP ON, LOAD,HZ') # turn on output
````
